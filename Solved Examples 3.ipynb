```python
from collections import defaultdict as dd
def msbfs(sources, root):
    global graph, visited, path, n, values
    max_distict = -1
    visited = [dd(int) for _ in range(n)]
    path = [[[] for _ in range(n)] for _ in range(n)]
    for i in sources:
        path[i][i].append(i)
        visited[i][i] = True
        visited[i][root] = True
    q = []
    for i in sources:
        q.append([i, -1, i])
    for node, parent, origin in q:
        for neigh in graph[node]:
            if not visited[origin][neigh]:
                path[origin][neigh] = path[origin][node] + [neigh]
                visited[origin][neigh] = True
                q.append([neigh, node, origin])
            else:
                if not parent == neigh:
                    path_origin2root = path[origin][node]+[neigh]
                    #print(path_origin2root)
                    total_path_val = set([values[i] for i in path_origin2root])
                    max_distict = max(max_distict, len(total_path_val))
                    
    return max_distict
                
```

```python

class SegmentTree():   
    def __init__(self, arr, func, fid=0):
        self.func = func
        self.tree = None
        self.f_identity = fid
        self.length = len(arr)
        self._build_tree(arr)
    def _build_tree(self, arr):            
        self.tree = [None]*(self.length - 1) + arr
        for i in range(self.length-2, -1, -1):
            self.tree[i] = self.func(self.tree[2*i+1], self.tree[2*i+2])
    def update(self, index, value):         
        index += (self.length - 1)
        self.tree[index] = value   
        index = ((index-1)//2)
        while index >= 0:   
            updated_internal_node_val = self.func(self.tree[2*index+1], self.tree[2*index+2])
            if updated_internal_node_val != self.tree[index]:
                self.tree[index] = updated_internal_node_val
            else: 
                break
            index = ((index-1)//2)
    def query(self, l, r):                   
        if l < 0 or r >= self.length:
            return None
        l += (self.length - 1)
        r += (self.length - 1)
        res = self.f_identity
        while r >= l:   
            if not (l & 1) :    
                res = self.func(res, self.tree[l])   
                l += 1
            if (r & 1) == 1 :     
                res = self.func(self.tree[r], res)
                r -= 1
            r = (r-1) // 2
            l = (l-1) // 2 
        return res
```
